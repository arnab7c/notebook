# Python 3 - Functions and OOPs | 4 | Classes and Objects 1 | Prob 2

Define the class comp that represents real and imaginary part of complex number.
for example, '2 + 3i', 2 is the real and 3 is the imaginary part
a. define a method add inside comp class which determines sum of two numbers and print "Sum of two complex Numbers: "
b. define a method sub inside comp class which determines sum of two numbers and print "Substraction of two complex Numbers: "

```python
import math
import os
import random
import re
import sys

class comp:
    def __init__(self,real,imaginary):
        self.real = real
        self.imaginary = imaginary

        
    def add(self,comp):
        c1 = complex(self.real,self.imaginary)
        c2 = complex(comp.real,comp.imaginary)
        sums='{:.0f}'.format(c1+c2)
        sums=sums.replace("j","i").replace("(","").replace(")","")
        print("Sum of the two Complex numbers :"+sums)
        
        
    def sub(self,comp):
        c1 = complex(self.real,self.imaginary)
        c2 = complex(comp.real,comp.imaginary)
        subs='{0:.0f}'.format(c1-c2)
        subs=subs.replace("j","i").replace("(","").replace(")","")
        print("Substraction of the two Complex numbers :"+subs)

if __name__ == '__main__':
    
    real1 = int(4)
    img1 = int(5)
    
    real2 = int(4)
    img2 = int(5)
    
    p1 = comp(real1,img1)
    p2 = comp(real2,img2)
    
    p1.add(p2)
    p1.sub(p2)
```

Devide total assets within family members
Rule:-
1. Parent will always inherit 50% of the family's total asset
2. Son will inherit based on the input % -> Percentage_for_son
3. Daughter will inherit based on the input % -> Percentage_for_daughter

parent class is defined
a. Define son class inherited from parent
    i. inherit __init__ function and add variable Percentage_for_son
    ii. define method son_display which will print share of son, rounded of to 2 decimal places
b. Define daughter class inherited from parent
    i. inherit __init__ function and add variable Percentage_for_daughter
    ii. define method daughter_display which will print share of daughter, rounded of to 2 decimal places

```python
import math
import os
import random
import re
import sys

class parent:
    def __init__(self,total_asset):
        self.total_asset = total_asset

    def display(self):
        print("Total Asset Worth is "+str(self.total_asset)+" Million.")
        print("Share of Parents is "+str(round(self.total_asset/2,2))+" Million.")


class son(parent):
    def __init__(self,total_asset,Percentage_for_son):
        super().__init__(total_asset)
        self.Percentage_for_son = Percentage_for_son
        
    def son_display(self):
        share = self.total_asset * (self.Percentage_for_son/100)
        print("Share of Son is",round(share,2),"Million.")


class daughter(parent):
    def __init__(self,total_asset,Percentage_for_daughter):
        super().__init__(total_asset)
        self.Percentage_for_daughter = Percentage_for_daughter
        
    def daughter_display(self):
        share = self.total_asset * (self.Percentage_for_daughter/100)
        print("Share of Daughter is",round(share,2),"Million.")

if __name__ == '__main__':
    
    t = int(10)
    sp = int(24)
    dp = int(26)


    obj1 = parent(t)
    

    obj2 = son(t,sp)
    obj2.son_display()
    obj2.display()


    obj3 = daughter(t,dp)
    obj3.daughter_display()
    obj3.display()
    
    print(isinstance(obj2,parent))
    print(isinst
          ance(obj3,parent))

    print(isinstance(obj3,son))
    print(isinstance(obj2,daughter))
```
