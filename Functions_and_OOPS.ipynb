Define the class comp that represents real and imaginary part of complex number.
for example, '2 + 3i', 2 is the real and 3 is the imaginary part
a. define a method add inside comp class which determines sum of two numbers and print "Sum of two complex Numbers: "
b. define a method sub inside comp class which determines sum of two numbers and print "Substraction of two complex Numbers: "

```python
import math
import os
import random
import re
import sys

class comp:
    def __init__(self,real,imaginary):
        self.real = real
        self.imaginary = imaginary

        
    def add(self,comp):
        c1 = complex(self.real,self.imaginary)
        c2 = complex(comp.real,comp.imaginary)
        sums='{:.0f}'.format(c1+c2)
        sums=sums.replace("j","i").replace("(","").replace(")","")
        print("Sum of the two Complex numbers :"+sums)
        
        
    def sub(self,comp):
        c1 = complex(self.real,self.imaginary)
        c2 = complex(comp.real,comp.imaginary)
        subs='{0:.0f}'.format(c1-c2)
        subs=subs.replace("j","i").replace("(","").replace(")","")
        print("Substraction of the two Complex numbers :"+subs)

if __name__ == '__main__':
    
    real1 = int(4)
    img1 = int(5)
    
    real2 = int(4)
    img2 = int(5)
    
    p1 = comp(real1,img1)
    p2 = comp(real2,img2)
    
    p1.add(p2)
    p1.sub(p2)
```

# Python 3 - Functions and OOPs | 7 | Handling Exceptions 2

Write the function definition for FORLoop which  takes no parameter
    - Get an integer input from user as n
    - create a list of n integer from user as l1
    - print l1
    - create an iterator for l1 and assign it iter1
    - using next() on iter1, print all elements in the list
    - return iterator variable iter1

```python
import os
import random
import re
import sys

def FORLoop():
    n=int(input())
    l1=[]
    for i in range(n):
        v=int(input())
        l1.append(v)
    print(l1)
    iter1=iter(l1)
    for x in range(n):
        print(next(iter1))
        
    return(iter1)

if __name__ == '__main__':
    try:
        d = FORLoop()
        print(type(d))
        print(next(d))
  
    except StopIteration:
        print('Stop Iteration : No Next Element to fetch')
```

# Python 3 - Functions and OOPs | 8 | Handling Exceptions 3

This exception handling scenario deals with the exceptional cases arises in bank ATM
ATM takes 3 integer input one by one
    1. balance - current balance of the account
    2. choice - coice of purpose, deposit(1) & withdrawl(2)
    3. amount - amount of deposit or withdrawl
Write a function Bank_ATM which takes above 3 input as parameter
a. raise ValueError if balance < 500 & print message
b. raise user defined error if deposit amount < 2000
c. raise user defined error if balance < 500 after withdrawl
d. if any transaction happen print update balance

```python
import math
import os
import random
import re
import sys


class MinimumDepositError(Exception):
    def __str__(self):
        return("The Minimum amount of Deposit should be 2000.")

class MinimumBalanceError(Exception):
    def __str__(self):
        return("You cannot withdraw this amount due to Minimum Balance Policy")


def Bank_ATM(balance,choice,amount):   
    if balance < 500:
        raise ValueError("As per the Minimum Balance Policy, Balance must be at least 500")
            
    elif choice == 1: 
        if amount < 2000:
            raise MinimumDepositError
        else:
            balance += amount
            print("Updated Balance Amount: ",balance)
            
    elif choice == 2:
        if balance - amount < 500:
            raise MinimumBalanceError
        else:
            balance -= amount
            print("Updated Balance Amount: ",balance)

if __name__ == '__main__':
    
    bal = int("1500")
    ch = int(1)
    amt = int(1000)
    
    try:
        Bank_ATM(bal,ch,amt)
    
    
    except ValueError as e:
        print(e)
    except MinimumDepositError as e:
        print(e)
    except MinimumBalanceError as e:
        print(e)
```
